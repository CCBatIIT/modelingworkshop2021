```python
import numpy as np
import scipy.integrate
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
xmin = -4
xmax = 4
x = np.linspace(xmin,xmax,500)

def u(x):
  return (x-2)*(x+2)*(x*x)
Z = scipy.integrate.quadrature(lambda x: np.exp(-u(x)),xmin,xmax)[0]

plt.figure(1, figsize=(8,6))
plt.plot(x,-np.log(np.exp(-u(x))/Z),'-')
plt.ylim([0,10])
plt.title('Energy landscape')

sigma = 2
plt.plot(x,-np.log(np.exp(-x*x/(2*sigma))/np.sqrt(2*np.pi)/sigma),'-.')
plt.plot(x,-np.log(np.exp(-(x-2)*(x-2)/(2*sigma))/np.sqrt(2*np.pi)/sigma),'--')

plt.legend(['Target','Okay for sampling','Bad for sampling'])
# plt.plot(x,np.exp(-(x-3)*(x-3)),'..')

plt.figure(2, figsize=(8,6))
plt.plot(x,np.exp(-u(x))/Z,'-')
plt.title('Probability density')

sigma = 2
plt.plot(x,np.exp(-x*x/(2*sigma))/np.sqrt(2*np.pi)/sigma,'-.')
plt.plot(x,np.exp(-(x-2)*(x-2)/(2*sigma))/np.sqrt(2*np.pi)/sigma,'--')

plt.legend(['Target','Okay for sampling','Bad for sampling'])
# plt.plot(x,np.exp(-(x-3)*(x-3)),'..')
```
