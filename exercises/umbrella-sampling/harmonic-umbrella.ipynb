```python
import numpy as np
import scipy.integrate
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
xmin = -4
xmax = 4
x = np.linspace(xmin,xmax,500)

def u(x):
  return (x-2)*(x+2)*(x*x)

def u_bias(x):
  return 2*x*x

Z = scipy.integrate.quadrature(lambda x: np.exp(-u(x)),xmin,xmax)[0]

plt.figure(1, figsize=(8,6))
plt.plot(x,u(x),'-')
plt.plot(x,u_bias(x),'-.')
plt.plot(x,u(x)+u_bias(x),'--')
plt.ylim([-5,10])
plt.title('Energy landscape')

plt.legend(['Target','Harmonic bias','Total PMF'])
```
